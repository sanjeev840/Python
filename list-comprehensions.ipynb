# List Comprehensions

```python
numbers = [1, 13, 4, 5, 63, 100]
numbers
```

```python
new_numbers = []
new_numbers
```

```python
for n in numbers:
    new_numbers.append(n * 2)

new_numbers
```

```python
new_numbers1 = [n * 3 for n in numbers]
new_numbers1
```

```python
for i in range(2):
    for j in range(5):
        print(i+j, end = " ")

```

```python
list_comprehension_1 = [i+j for i in range(2) for j in range(5)]
list_comprehension_1
```

```python
type(list_comprehension_1[1])
```

```python
list_comprehension_2 = [ [i+j for i in range(2)] for j in range(5)]
list_comprehension_2
```

```python
type(list_comprehension_2[1])
```

```python
for n in range(1, 11):
    if (n % 2 != 0):
        print(n ** 3, end =" ")
```

```python
[n ** 3 for n in range(1, 11) if n %2 != 0]
```

```python
[n ** 3 if (n %2 != 0) else "even" for n in range(1, 11) ]
```
